```python
# ============================================================
#  DAILY Cross-Asset HMM + Templates + MVO
# STRICTLY CAUSAL Expanding Window:
#   - At OOS date d: fit/score using data up to d-1 ONLY
#   - Regime moments use FORWARD returns: r_{s+1} grouped by regime z_s

#   (1) NO BIC-once.  K is selected periodically via predictive score (rolling validation).
#   (2) NO emergence-only monotone K. K can go up/down based on predictive score.
#   (3) NO Hungarian / linear assignment label matching.
#       Instead: persistent TEMPLATE identities tracked via Wasserstein distance + EMA updates.
#
# EXTRA: prints K at EVERY iteration (each OOS day).
# ============================================================

import numpy as np
import pandas as pd
import yfinance as yf
import cvxpy as cp
import matplotlib.pyplot as plt

from hmmlearn.hmm import GaussianHMM
from sklearn.covariance import LedoitWolf
from tqdm import tqdm

np.random.seed(42)

# ------------------------------------------------------------
# 1) Asset universe
# ------------------------------------------------------------
tickers = {
    "SPX": "^GSPC",
    "BOND": "IEF",
    "GOLD": "GLD",
    "OIL": "USO",
    "USD": "UUP"
}
asset_names = list(tickers.keys())
N = len(asset_names)

# ------------------------------------------------------------
# 2) Download DAILY data
# ------------------------------------------------------------
END_DATE = pd.Timestamp.today().strftime("%Y-%m-%d")


# After yf.download(... )["Close"]
prices = yf.download(
    list(tickers.values()),
    start="2005-01-01",
    end=END_DATE,
    auto_adjust=True,
    progress=False
)["Close"]

# ✅ Hardcoded rename: Yahoo symbol -> your name
prices = prices.rename(columns={
    "^GSPC": "SPX",
    "IEF":   "BOND",
    "GLD":   "GOLD",
    "USO":   "OIL",
    "UUP":   "USD",
})

# (optional) enforce column order
prices = prices[["SPX", "BOND", "GOLD", "OIL", "USD"]]

# clean
prices = prices.dropna(how="any")

# ------------------------------------------------------------
# 3) DAILY log returns
# ------------------------------------------------------------
returns = np.log(prices).diff().dropna()

# ------------------------------------------------------------
# 4) Train / Test split (NO OVERLAP)
# ------------------------------------------------------------
split_date = pd.Timestamp("2023-06-03")
returns_train = returns.loc[returns.index <  split_date]
returns_test  = returns.loc[returns.index >= split_date]

# ------------------------------------------------------------
# 5) Feature construction (ret + vol + mom)
# ------------------------------------------------------------
def build_features(ret_df: pd.DataFrame, vol_window: int = 60, mom_window: int = 20) -> pd.DataFrame:
    vol = ret_df.rolling(vol_window).std()
    mom = ret_df.rolling(mom_window).mean()
    feats = pd.concat([ret_df, vol, mom], axis=1)
    feats.columns = (
        [f"ret_{c}" for c in ret_df.columns] +
        [f"vol_{c}" for c in ret_df.columns] +
        [f"mom_{c}" for c in ret_df.columns]
    )
    return feats.dropna()

# ------------------------------------------------------------
# 6) MVO (same as KNN)
# ------------------------------------------------------------
def solve_mvo(mu, Sigma, w_prev, lam=5.0, tc=0.0002, w_max=0.6, solver="OSQP"):
    mu = np.asarray(mu).reshape(-1)
    n = mu.shape[0]

    Sigma = np.asarray(Sigma)
    Sigma = 0.5 * (Sigma + Sigma.T) + 1e-8 * np.eye(n)

    w = cp.Variable(n)
    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - tc * cp.norm1(w - w_prev)
    )

    constraints = [
        cp.sum(w) == 1,
        w >= 0,
        cp.norm_inf(w) <= w_max
    ]

    prob = cp.Problem(objective, constraints)

    try:
        prob.solve(solver=getattr(cp, solver), verbose=False)
        if w.value is None:
            prob.solve(solver=cp.SCS, verbose=False)
    except Exception:
        prob.solve(solver=cp.SCS, verbose=False)

    if w.value is None:
        return w_prev.copy()

    w_sol = np.asarray(w.value).reshape(-1)
    w_sol = np.maximum(w_sol, 0.0)
    s = w_sol.sum()
    if s <= 1e-12:
        return w_prev.copy()
    return w_sol / s

# ------------------------------------------------------------
# 7) STRICTLY CAUSAL regime moments using FORWARD returns
# ------------------------------------------------------------
def compute_regime_moments_forward(ret_align: pd.DataFrame,
                                   regimes_s: np.ndarray,
                                   K: int,
                                   min_obs: int = 60):
    """
    STRICTLY CAUSAL regime moments:
      - regimes_s are regime labels for dates s (aligned with features at s)
      - returns used are FORWARD returns r_{s+1} grouped by regime z_s

    Inputs:
      ret_align: returns indexed by dates included in X (same index as features)
      regimes_s: array length len(ret_align)-1 (we drop last because no forward label)
    """
    r_fwd = ret_align.shift(-1).dropna()  # length len(ret_align)-1
    r_fwd = r_fwd.iloc[:len(regimes_s)]   # safety

    MU = np.zeros((K, N))
    SIG = np.zeros((K, N, N))

    for k in range(K):
        rk = r_fwd.iloc[(regimes_s == k)]
        if len(rk) < min_obs:
            MU[k, :] = 0.0
            SIG[k, :, :] = np.eye(N)
        else:
            MU[k, :] = rk.mean().values
            SIG[k, :, :] = LedoitWolf().fit(rk.values).covariance_
    return MU, SIG

# ------------------------------------------------------------
# 8) Wasserstein-2 distance between Gaussians (stable via eigen)
# ------------------------------------------------------------
def _symmetrize(A):
    return 0.5 * (A + A.T)

def _sqrtm_psd(A, eps=1e-12):
    A = _symmetrize(A)
    w, V = np.linalg.eigh(A)
    w = np.maximum(w, eps)
    return V @ np.diag(np.sqrt(w)) @ V.T

def w2_gaussian(mu1, S1, mu2, S2, eps=1e-12):
    """
    W2^2(N(mu1,S1), N(mu2,S2)) = ||mu1-mu2||^2 + Tr(S1+S2 - 2*(S2^{1/2} S1 S2^{1/2})^{1/2})
    Returns W2 (not squared).
    """
    mu1 = np.asarray(mu1).reshape(-1)
    mu2 = np.asarray(mu2).reshape(-1)
    S1 = _symmetrize(np.asarray(S1))
    S2 = _symmetrize(np.asarray(S2))

    dm2 = float(np.sum((mu1 - mu2) ** 2))

    S2_sqrt = _sqrtm_psd(S2, eps=eps)
    M = S2_sqrt @ S1 @ S2_sqrt
    M_sqrt = _sqrtm_psd(M, eps=eps)

    tr_term = float(np.trace(S1 + S2 - 2.0 * M_sqrt))
    tr_term = max(tr_term, 0.0)
    return float(np.sqrt(dm2 + tr_term))

# ------------------------------------------------------------
# 9) Template tracking (no assignment optimization)
# ------------------------------------------------------------
def map_components_to_templates(MU_K, SIG_K, MU_tpl, SIG_tpl):
    """
    For each component k, map to nearest template g by Wasserstein distance.
    Returns:
      map_k_to_g: list length K
      dist_k: list length K (best distances)
    """
    K = MU_K.shape[0]
    G = MU_tpl.shape[0]
    map_k_to_g = []
    dist_k = []
    for k in range(K):
        dists = [w2_gaussian(MU_tpl[g], SIG_tpl[g], MU_K[k], SIG_K[k]) for g in range(G)]
        g_best = int(np.argmin(dists))
        map_k_to_g.append(g_best)
        dist_k.append(float(dists[g_best]))
    return map_k_to_g, dist_k

def spawn_template_if_needed(MU_K, SIG_K, pK, MU_tpl, SIG_tpl, spawn_thresh, G_max):
    """
    Optional: if a component is far from all templates and we have room, spawn a new template.
    We pick the component with highest posterior mass among those exceeding threshold.
    """
    if MU_tpl is None:
        return MU_K.copy(), SIG_K.copy()

    G = MU_tpl.shape[0]
    if G >= G_max:
        return MU_tpl, SIG_tpl

    map_k_to_g, dist_k = map_components_to_templates(MU_K, SIG_K, MU_tpl, SIG_tpl)
    dist_k = np.asarray(dist_k)
    candidates = np.where(dist_k > spawn_thresh)[0]
    if len(candidates) == 0:
        return MU_tpl, SIG_tpl

    # choose the far component with highest posterior mass
    k_new = int(candidates[np.argmax(np.asarray(pK)[candidates])])
    MU_tpl_new = np.vstack([MU_tpl, MU_K[k_new][None, :]])
    SIG_tpl_new = np.concatenate([SIG_tpl, SIG_K[k_new][None, :, :]], axis=0)
    return MU_tpl_new, SIG_tpl_new

def update_templates_ema(MU_tpl, SIG_tpl, MU_K, SIG_K, pK, map_k_to_g, eta=0.05):
    """
    EMA update: each template g gets posterior-weighted average of components mapped to g.
    """
    MU_tpl = MU_tpl.copy()
    SIG_tpl = SIG_tpl.copy()
    G = MU_tpl.shape[0]
    K = MU_K.shape[0]

    pK = np.asarray(pK).reshape(-1)
    for g in range(G):
        ks = [k for k in range(K) if map_k_to_g[k] == g]
        if len(ks) == 0:
            continue
        w = pK[ks]
        ws = float(w.sum())
        if ws <= 1e-12:
            continue
        w = w / ws

        mu_bar = (w[:, None] * MU_K[ks]).sum(axis=0)
        S_bar = np.zeros((N, N))
        for kk, wk in zip(ks, w):
            S_bar += wk * SIG_K[kk]

        MU_tpl[g] = (1 - eta) * MU_tpl[g] + eta * mu_bar
        SIG_tpl[g] = _symmetrize((1 - eta) * SIG_tpl[g] + eta * S_bar) + 1e-8 * np.eye(N)

    return MU_tpl, SIG_tpl

def aggregate_template_posteriors(pK, map_k_to_g, G):
    pK = np.asarray(pK).reshape(-1)
    pG = np.zeros(G)
    for k, g in enumerate(map_k_to_g):
        pG[g] += pK[k]
    s = pG.sum()
    if s > 1e-12:
        pG /= s
    return pG

# ------------------------------------------------------------
# 10) Predictive K selection (strictly causal)
# ------------------------------------------------------------
def select_K_predictive(X_all, K_candidates, L_val=252, n_iter=300, random_state=42, lamK=1.0):
    """
    Strictly causal within-history selection:
      - Fit on X_fit = X_all[:-L_val]
      - Score on X_val = X_all[-L_val:]
    Use hmm.score(X_val) as predictive log score proxy.
    Penalize complexity by lamK * K.
    """
    if len(X_all) <= (L_val + 50):
        # too little data: pick smallest K
        return int(min(K_candidates))

    X_fit = X_all[:-L_val]
    X_val = X_all[-L_val:]

    best_score = -np.inf
    best_K = int(min(K_candidates))

    for K in K_candidates:
        try:
            hmm = GaussianHMM(
                n_components=int(K),
                covariance_type="full",
                n_iter=n_iter,
                random_state=random_state
            )
            hmm.fit(X_fit)
            predll = float(hmm.score(X_val))  # proxy for predictive score
            score = predll - lamK * float(K)
            if score > best_score:
                best_score = score
                best_K = int(K)
        except Exception:
            continue

    return best_K

# ------------------------------------------------------------
# 11) Settings
# ------------------------------------------------------------
LOOKBACK = 60

LAM = 5.0
TC  = 0.0002
W_MAX = 0.6

MIN_REGIMES = 5
MAX_REGIMES = 6

# K selection controls
F_K = 5               # select K every 5 trading days (weekly)
L_VAL = 252           # validation length (~1y of trading days) inside history
LAM_K = 1.0           # complexity penalty weight (penalize larger K)

# Template controls
G_MAX = 8             # allow up to this many templates (can be >= MAX_REGIMES)
ETA_TPL = 0.05        # EMA learning rate for templates
SPAWN_THRESH = 2.5    # W2 distance threshold to spawn a new template (tune)

# ------------------------------------------------------------
# 12) STRICTLY CAUSAL DAILY expanding-window HMM backtest (Commercial V2.0)
# ------------------------------------------------------------
weights_oos, pnl_oos, dates_oos = [], [], []
K_history = []

# Template label history (robust label = template id)
tpl_label_history = []
tpl_prob_history = []
tpl_count_history = []

w_prev = np.zeros(N)
K_curr = MIN_REGIMES

# Templates (persistent identities)
MU_tpl = None
SIG_tpl = None

print("Running COMMERCIAL V2.0: Predictive-K + Template Tracking + MVO (STRICTLY CAUSAL)\n")

for t_i, date in enumerate(tqdm(returns_test.index, desc="Backtest progress", ncols=100)):
    # STRICTLY CAUSAL: history ends at date-1
    ret_hist = pd.concat([returns_train, returns_test.loc[returns_test.index < date]]).sort_index()
    ret_hist = ret_hist[~ret_hist.index.duplicated(keep="last")]

    Xdf = build_features(ret_hist, vol_window=60, mom_window=20)
    if len(Xdf) < max(LOOKBACK, 2):
        w_t = w_prev.copy()
        pnl = float(np.dot(w_t, returns.loc[date].values))
        weights_oos.append(w_t); pnl_oos.append(pnl); dates_oos.append(date)
        K_history.append(np.nan)
        tpl_label_history.append(np.nan)
        tpl_prob_history.append(np.nan)
        tpl_count_history.append(0)
        continue

    idx = Xdf.index
    ret_align = ret_hist.loc[idx]
    X = Xdf.values

    if len(X) < 2:
        w_t = w_prev.copy()
        pnl = float(np.dot(w_t, returns.loc[date].values))
        weights_oos.append(w_t); pnl_oos.append(pnl); dates_oos.append(date)
        K_history.append(K_curr)
        tpl_label_history.append(np.nan)
        tpl_prob_history.append(np.nan)
        tpl_count_history.append(0 if MU_tpl is None else MU_tpl.shape[0])
        continue

    # --------------------------------------------------------
    # (A) Periodic predictive K selection (weekly)
    # --------------------------------------------------------


    if (t_i % F_K == 0) or (t_i == 0):

    # MONOTONE K: only consider K >= current K
       K_candidates = list(range(K_curr, MAX_REGIMES + 1))

       K_candidate = select_K_predictive(
          X_all=X,
          K_candidates=K_candidates,
          L_val=L_VAL,
          n_iter=300,
          random_state=42,
          lamK=LAM_K
       )

    # Safety: enforce non-decreasing K
    K_curr = max(K_curr, K_candidate)

    # PRINT K at EVERY iteration
    print(f"{pd.Timestamp(date).date()}  K = {K_curr}")

    # --------------------------------------------------------
    # (B) Fit HMM at chosen K on ALL history up to date-1
    # --------------------------------------------------------
    hmmK = GaussianHMM(n_components=K_curr, covariance_type="full", n_iter=300, random_state=42)
    hmmK.fit(X)

    probsK = hmmK.predict_proba(X)
    pK = probsK[-1].copy()
    zK_raw = np.argmax(probsK, axis=1)

    # Regime moments with forward returns r_{s+1} grouped by z_s
    z_s = zK_raw[:-1]
    MU_K, SIG_K = compute_regime_moments_forward(ret_align, z_s, K_curr, min_obs=60)

    # --------------------------------------------------------
    # (C) Initialize / spawn templates, then map components -> templates
    # --------------------------------------------------------
    if MU_tpl is None or SIG_tpl is None:
        # initialize templates directly from current components
        MU_tpl = MU_K.copy()
        SIG_tpl = SIG_K.copy()
    else:
        # optionally spawn a new template if a component is "far" from all templates
        MU_tpl, SIG_tpl = spawn_template_if_needed(
            MU_K, SIG_K, pK,
            MU_tpl, SIG_tpl,
            spawn_thresh=SPAWN_THRESH,
            G_max=G_MAX
        )

    # map each component to nearest template
    map_k_to_g, _ = map_components_to_templates(MU_K, SIG_K, MU_tpl, SIG_tpl)
    G = MU_tpl.shape[0]
    pG = aggregate_template_posteriors(pK, map_k_to_g, G)

    # update templates with EMA
    MU_tpl, SIG_tpl = update_templates_ema(
        MU_tpl, SIG_tpl,
        MU_K, SIG_K,
        pK, map_k_to_g,
        eta=ETA_TPL
    )

    # --------------------------------------------------------
    # (D) Template mixture moments + MVO
    # --------------------------------------------------------
    mu_t = MU_tpl.T @ pG
    Sigma_t = np.tensordot(pG, SIG_tpl, axes=(0, 0))

    w_t = solve_mvo(mu_t, Sigma_t, w_prev, lam=LAM, tc=TC, w_max=W_MAX, solver="OSQP")

    pnl = float(np.dot(w_t, returns.loc[date].values))

    weights_oos.append(w_t)
    pnl_oos.append(pnl)
    dates_oos.append(date)
    K_history.append(K_curr)

    # robust label = template hard label (argmax over pG)
    g_hat = int(np.argmax(pG)) if np.all(np.isfinite(pG)) else np.nan
    tpl_label_history.append(g_hat)
    tpl_prob_history.append(float(np.max(pG)) if np.all(np.isfinite(pG)) else np.nan)
    tpl_count_history.append(G)

    w_prev = w_t

# ------------------------------------------------------------
# 13) Performance evaluation
# ------------------------------------------------------------
pnl_oos = pd.Series(pnl_oos, index=pd.to_datetime(dates_oos)).sort_index()
cum_pnl = pnl_oos.cumsum()

sharpe_ratio = np.sqrt(252) * pnl_oos.mean() / (pnl_oos.std() + 1e-12)
max_dd = (cum_pnl - cum_pnl.cummax()).min()

print("\n========== COMMERCIAL V2.0 HMM+TEMPLATES + MVO PERFORMANCE ==========")
print(f"OOS Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Max Drawdown: {max_dd:.2%}")

# ------------------------------------------------------------
# 14) Diagnostics
# ------------------------------------------------------------
W = pd.DataFrame(weights_oos, index=pnl_oos.index, columns=asset_names)
dW = W.diff().dropna()

turnover = 0.5 * dW.abs().sum(axis=1)
n_eff = 1.0 / (W.pow(2).sum(axis=1) + 1e-12)

print("\n========== REBALANCE FREQUENCY ==========")
print("Rebalance rule: DAILY")
print(f"Average daily turnover: {turnover.mean():.4f}")
print(f"Median daily turnover : {turnover.median():.4f}")
print(f"95% turnover quantile : {turnover.quantile(0.95):.4f}")
print(f"Fraction of days with >1% turnover: {(turnover > 0.01).mean():.2%}")
print(f"Fraction of days with >5% turnover: {(turnover > 0.05).mean():.2%}")

alloc_summary = pd.DataFrame({
    "Average Weight": W.mean(),
    "Weight Volatility": W.std(),
    "Time >10%": (W > 0.10).mean(),
    "Avg |Δw|": dW.abs().mean().reindex(asset_names)
})

print("\n========== AVERAGE ALLOCATION ==========")
print(alloc_summary)

print("\n========== CONCENTRATION ==========")
print(f"Average N_eff: {n_eff.mean():.2f}")
print(f"Median  N_eff: {n_eff.median():.2f}")

# ------------------------------------------------------------
# 15) Plots
# ------------------------------------------------------------
plt.figure(figsize=(10, 5))
plt.plot(cum_pnl, label="Commercial V2.0 OOS Cumulative PnL")
plt.title("Commercial V2.0 — Cumulative PnL")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(turnover, label="Daily Turnover (0.5*L1)")
plt.title("Turnover Over Time")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(n_eff, label="Effective # Positions (N_eff)")
plt.title("Portfolio Concentration Over Time")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(11, 5))
plt.stackplot(W.index, [W[c].values for c in W.columns], labels=W.columns)
plt.title("Portfolio Weights (Stacked)")
plt.legend(loc="upper left", ncol=5)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 3))
plt.plot(pd.Series(K_history, index=pnl_oos.index), linewidth=1)
plt.title("Selected K Over Time (Predictive Selection)")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 3))
plt.plot(pd.Series(tpl_count_history, index=pnl_oos.index), linewidth=1)
plt.title("Number of Templates Over Time")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 3))
plt.plot(pd.Series(tpl_label_history, index=pnl_oos.index), linewidth=1)
plt.title("ROBUST Label Over Time (Template hard label)")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 3))
plt.plot(pd.Series(tpl_prob_history, index=pnl_oos.index), linewidth=1)
plt.title("Max Template Posterior Over Time (Confidence proxy)")
plt.grid(True)
plt.show()
```

```python
# ============================================================
# ADD-ON: Sharpe / performance BY TEMPLATE REGIME (Commercial V2.0)
# Works with your existing backtest outputs:
#   - pnl_oos : pd.Series (daily portfolio returns), index = dates
#   - tpl_label_history : list/array of template hard labels per OOS date
# Optionally:
#   - asset_names : list[str]
#   - returns : pd.DataFrame of asset daily log returns (same index universe)
#   - W : pd.DataFrame of weights (index aligned to pnl_oos)
# ============================================================

import numpy as np
import pandas as pd

# -----------------------------
# 0) Safety checks
# -----------------------------
assert isinstance(pnl_oos, pd.Series), "pnl_oos must be a pd.Series of daily portfolio returns"
assert len(pnl_oos) == len(tpl_label_history), "tpl_label_history length must match pnl_oos length"

# Align regime labels to pnl index
regime = pd.Series(tpl_label_history, index=pnl_oos.index, name="template_regime")

df = pd.DataFrame({
    "pnl": pnl_oos.astype(float),
    "regime": regime.astype("float")
}).dropna()

# -----------------------------
# 1) Helpers
# -----------------------------
TRADING_DAYS = 252

def ann_mean(x: pd.Series) -> float:
    return float(x.mean() * TRADING_DAYS)

def ann_vol(x: pd.Series) -> float:
    return float(x.std(ddof=1) * np.sqrt(TRADING_DAYS))

def ann_sharpe(x: pd.Series) -> float:
    return float(np.sqrt(TRADING_DAYS) * x.mean() / (x.std(ddof=1) + 1e-12))

def hit_rate(x: pd.Series) -> float:
    return float((x > 0).mean())

def max_drawdown_from_returns(r: pd.Series) -> float:
    # treat r as arithmetic daily returns; cumulative wealth
    wealth = (1.0 + r.fillna(0.0)).cumprod()
    dd = wealth / wealth.cummax() - 1.0
    return float(dd.min())

# -----------------------------
# 2) Portfolio performance by template regime
# -----------------------------
grp = df.groupby("regime")["pnl"]

regime_report = pd.DataFrame({
    "Days": grp.size().astype(int),
    "Ann Mean": grp.apply(ann_mean),
    "Ann Vol": grp.apply(ann_vol),
    "Sharpe": grp.apply(ann_sharpe),
    "Hit Rate": grp.apply(hit_rate),
    "Max DD (within-regime)": grp.apply(max_drawdown_from_returns),
}).sort_index()

# Add occupancy
regime_report["Frac of Sample"] = regime_report["Days"] / regime_report["Days"].sum()

print("\n========== PORTFOLIO PERFORMANCE BY TEMPLATE REGIME ==========")
print(regime_report)

# -----------------------------
# 3) OPTIONAL: Asset performance by template regime
#    Requires: returns (pd.DataFrame), asset_names (list of columns)
# -----------------------------
if "returns" in globals() and isinstance(returns, pd.DataFrame):
    # align asset returns to pnl index (OOS dates)
    R = returns.loc[df.index].copy()
    if "asset_names" in globals():
        R = R[asset_names].copy()

    asset_stats = {}
    for col in R.columns:
        gcol = pd.concat([R[col], df["regime"]], axis=1).dropna()
        ggrp = gcol.groupby("regime")[col]
        asset_stats[col] = pd.DataFrame({
            "Ann Mean": ggrp.apply(ann_mean),
            "Ann Vol": ggrp.apply(ann_vol),
            "Sharpe": ggrp.apply(ann_sharpe),
        }).sort_index()

    # Combine into multiindex table: (asset, metric)
    asset_report = pd.concat(asset_stats, axis=1)
    print("\n========== ASSET PERFORMANCE BY TEMPLATE REGIME (ANNUALIZED) ==========")
    print(asset_report)

# -----------------------------
# 4) OPTIONAL: Average weights by template regime
#    Requires: W (pd.DataFrame of weights aligned to pnl_oos)
# -----------------------------
if "W" in globals() and isinstance(W, pd.DataFrame):
    W_aligned = W.loc[df.index].copy()
    W_aligned["regime"] = df["regime"].values
    avg_w_by_regime = W_aligned.groupby("regime").mean().drop(columns=["regime"], errors="ignore")
    print("\n========== AVERAGE WEIGHTS BY TEMPLATE REGIME ==========")
    print(avg_w_by_regime.sort_index())
```

```python
# ============================================================
#  Cross-Asset KNN + Mean-Variance Optimization (MVO)
#  Expanding Window | STRICTLY CAUSAL OOS Backtest (NO LEAKAGE)
#  - No train/test overlap
#  - Features at day t predict returns at day t+1 (no same-day label leakage)
#  - Date-aligned mapping between features and returns (no slicing mismatch)
# ============================================================

import numpy as np
import pandas as pd
import yfinance as yf
import cvxpy as cp
import matplotlib.pyplot as plt

from sklearn.neighbors import NearestNeighbors
from sklearn.covariance import LedoitWolf
from tqdm import tqdm

np.random.seed(42)

# ------------------------------------------------------------
# 1) Asset universe
# ------------------------------------------------------------
tickers = {
    "SPX": "^GSPC",
    "BOND": "IEF",
    "GOLD": "GLD",
    "OIL": "USO",
    "USD": "UUP"
}
asset_names = list(tickers.keys())

# ------------------------------------------------------------
# 2) Download DAILY data
# ------------------------------------------------------------
END_DATE = pd.Timestamp.today().strftime("%Y-%m-%d")

prices = yf.download(
    list(tickers.values()),
    start="2005-01-01",
    end=END_DATE,
    auto_adjust=True,
    progress=False
)["Close"]

prices.columns = asset_names
prices = prices.dropna(how="any")

# ------------------------------------------------------------
# 3) DAILY log returns
# ------------------------------------------------------------
returns = np.log(prices).diff().dropna()

# ------------------------------------------------------------
# 4) Train / Test split (NO OVERLAP)
# ------------------------------------------------------------
split_date = pd.Timestamp("2023-06-01")
returns_train = returns.loc[returns.index < split_date]
returns_test  = returns.loc[returns.index >= split_date]

# ------------------------------------------------------------
# 5) Feature construction (DAILY)
#   - raw returns
#   - rolling vol
#   - rolling momentum (mean return)
# ------------------------------------------------------------
def build_features(ret_df: pd.DataFrame, vol_window: int = 60, mom_window: int = 20) -> pd.DataFrame:
    vol = ret_df.rolling(vol_window).std()
    mom = ret_df.rolling(mom_window).mean()
    feats = pd.concat([ret_df, vol, mom], axis=1)
    feats.columns = (
        [f"ret_{c}" for c in ret_df.columns] +
        [f"vol_{c}" for c in ret_df.columns] +
        [f"mom_{c}" for c in ret_df.columns]
    )
    return feats.dropna()

# ------------------------------------------------------------
# 6) Mean-Variance Optimization with scaled transaction costs
#    objective: mu'w - lam*w'Sigma*w - tc*||w-w_prev||_1
# ------------------------------------------------------------
def solve_mvo(mu, Sigma, w_prev, lam=5.0, tc=0.0002, w_max=0.6, solver="OSQP"):
    mu = np.asarray(mu).reshape(-1)
    n = mu.shape[0]

    Sigma = np.asarray(Sigma)
    Sigma = 0.5 * (Sigma + Sigma.T) + 1e-8 * np.eye(n)

    w = cp.Variable(n)

    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - tc * cp.norm1(w - w_prev)
    )

    constraints = [
        cp.sum(w) == 1,
        w >= 0,
        cp.norm_inf(w) <= w_max
    ]

    prob = cp.Problem(objective, constraints)

    try:
        prob.solve(solver=getattr(cp, solver), verbose=False)
        if w.value is None:
            prob.solve(solver=cp.SCS, verbose=False)
    except Exception:
        prob.solve(solver=cp.SCS, verbose=False)

    if w.value is None:
        return w_prev.copy()

    w_sol = np.asarray(w.value).reshape(-1)
    w_sol = np.maximum(w_sol, 0.0)
    s = w_sol.sum()
    if s <= 1e-12:
        return w_prev.copy()
    return w_sol / s

# ------------------------------------------------------------
# 7) Settings
# ------------------------------------------------------------
K_NEIGHBORS = 40
VOL_WINDOW = 60
MOM_WINDOW = 20
REB_FREQ = 5

MU_TRIGGER_L1 = 0.002
SIG_TRIGGER_FROB = 0.20

LAM = 5.0
TC  = 0.0002
W_MAX = 0.6

# ------------------------------------------------------------
# 8) STRICTLY CAUSAL expanding-window KNN backtest
#
# Key idea:
#   - At OOS date d (in returns_test):
#       signal is built from features at d-1
#       targets for neighbors are forward returns r_{s+1}
#       PnL uses r_d (return realized on d)
#
# Implementation:
#   - Build history up to d-1 (exclude current day return)
#   - Build features X over that history
#   - Take x_t = X at last feature date (typically d-1)
#   - Fit KNN on earlier feature rows
#   - For those training feature dates s, use forward returns r_{s+1}
# ------------------------------------------------------------
weights_oos, pnl_oos, dates_oos = [], [], []
rebalanced_flags = []

w_prev = np.zeros(len(asset_names))
mu_prev = None
Sigma_prev = None

print("\nRunning STRICTLY CAUSAL KNN + MVO backtest...\n")

for date in tqdm(returns_test.index, desc="Backtest progress", ncols=100):
    # History ends at date-1 (exclude today's return from all signal estimation)
    hist_end = date
    ret_hist = pd.concat([returns_train, returns_test.loc[returns_test.index < hist_end]])

    # Need enough history for features + forward labels
    X_all = build_features(ret_hist, vol_window=VOL_WINDOW, mom_window=MOM_WINDOW)
    if len(X_all) < (K_NEIGHBORS + 2):
        w_t = w_prev.copy()
        pnl = float(np.dot(w_t, returns.loc[date].values))
        weights_oos.append(w_t); pnl_oos.append(pnl); dates_oos.append(date)
        rebalanced_flags.append(False)
        continue

    # Current signal uses last available feature row (typically date-1)
    x_date = X_all.index[-1]
    x_t = X_all.loc[[x_date]].values  # shape (1, p)

    # Training feature rows exclude the last row (so we don't include the "query" itself)
    X_train = X_all.iloc[:-1].copy()

    # Align returns to feature dates, then create forward (next-day) returns labels
    ret_aligned = ret_hist.loc[X_train.index]          # returns at s
    ret_fwd = ret_aligned.shift(-1).dropna()           # returns at s+1

    # Keep only feature rows that have a forward label
    X_train = X_train.loc[ret_fwd.index]

    # If after dropping we don't have enough samples, hold
    if len(X_train) < (K_NEIGHBORS + 1):
        w_t = w_prev.copy()
        pnl = float(np.dot(w_t, returns.loc[date].values))
        weights_oos.append(w_t); pnl_oos.append(pnl); dates_oos.append(date)
        rebalanced_flags.append(False)
        continue

    # KNN on feature space
    knn = NearestNeighbors(n_neighbors=K_NEIGHBORS, metric="euclidean")
    knn.fit(X_train.values)
    _, idx = knn.kneighbors(x_t)

    # Neighbor forward returns (labels)
    neighbor_returns = ret_fwd.iloc[idx.flatten()]

    # Local conditional moments (for next-day returns)
    mu_t = neighbor_returns.mean().values
    Sigma_t = LedoitWolf().fit(neighbor_returns.values).covariance_

    # Rebalance decision (weekly schedule OR triggered)
    # Use x_date to decide schedule consistency (not strictly necessary)
    # We'll count OOS step index relative to first OOS date:
    oos_step = returns_test.index.get_loc(date)
    on_schedule = (oos_step % REB_FREQ == 0)

    trigger = False
    if mu_prev is not None and Sigma_prev is not None:
        mu_shift = np.sum(np.abs(mu_t - mu_prev))
        denom = np.linalg.norm(Sigma_prev, ord="fro") + 1e-12
        sig_shift = np.linalg.norm(Sigma_t - Sigma_prev, ord="fro") / denom
        trigger = (mu_shift > MU_TRIGGER_L1) or (sig_shift > SIG_TRIGGER_FROB)

    do_rebalance = on_schedule or trigger

    if do_rebalance:
        w_t = solve_mvo(mu_t, Sigma_t, w_prev, lam=LAM, tc=TC, w_max=W_MAX, solver="OSQP")
        mu_prev = mu_t.copy()
        Sigma_prev = Sigma_t.copy()
    else:
        w_t = w_prev.copy()

    # Realized PnL uses return at current OOS date (date)
    r_today = returns.loc[date].values
    pnl = float(np.dot(w_t, r_today))

    weights_oos.append(w_t)
    pnl_oos.append(pnl)
    dates_oos.append(date)
    rebalanced_flags.append(do_rebalance)

    w_prev = w_t

# ------------------------------------------------------------
# 9) Performance evaluation
# ------------------------------------------------------------
pnl_oos = pd.Series(pnl_oos, index=pd.to_datetime(dates_oos)).sort_index()
cum_pnl = pnl_oos.cumsum()

sharpe_ratio = np.sqrt(252) * pnl_oos.mean() / (pnl_oos.std() + 1e-12)
max_dd = (cum_pnl - cum_pnl.cummax()).min()

print("\n========== KNN + MVO PERFORMANCE (STRICTLY CAUSAL) ==========")
print(f"OOS Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Max Drawdown: {max_dd:.2%}")

# ------------------------------------------------------------
# 10) Rebalance / allocation diagnostics
# ------------------------------------------------------------
W = pd.DataFrame(weights_oos, index=pnl_oos.index, columns=asset_names)
dW = W.diff().dropna()

turnover = 0.5 * dW.abs().sum(axis=1)
n_eff = 1.0 / (W.pow(2).sum(axis=1) + 1e-12)

reb_series = pd.Series(rebalanced_flags, index=pnl_oos.index).astype(int)
reb_days = reb_series.sum()

print("\n========== REBALANCE FREQUENCY ==========")
print(f"Rebalance rule: every {REB_FREQ} trading days OR trigger")
print(f"Total rebalance days: {int(reb_days)} out of {len(reb_series)} ({reb_days/len(reb_series):.2%})")
print(f"Average daily turnover: {turnover.mean():.4f}")
print(f"Median daily turnover : {turnover.median():.4f}")
print(f"95% turnover quantile : {turnover.quantile(0.95):.4f}")
print(f"Fraction of days with >1% turnover: {(turnover > 0.01).mean():.2%}")
print(f"Fraction of days with >5% turnover: {(turnover > 0.05).mean():.2%}")

alloc_summary = pd.DataFrame({
    "Average Weight": W.mean(),
    "Weight Volatility": W.std(),
    "Time >10%": (W > 0.10).mean(),
    "Avg |Δw|": dW.abs().mean().reindex(asset_names)
})

print("\n========== AVERAGE ALLOCATION ==========")
print(alloc_summary)

print("\n========== CONCENTRATION ==========")
print(f"Average N_eff: {n_eff.mean():.2f}")
print(f"Median  N_eff: {n_eff.median():.2f}")

# ------------------------------------------------------------
# 11) Plots
# ------------------------------------------------------------
plt.figure(figsize=(10, 5))
plt.plot(cum_pnl, label="KNN + MVO OOS Cumulative PnL")
plt.title("KNN-Based Allocation (STRICTLY CAUSAL) — Cumulative PnL")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(turnover, label="Daily Turnover (0.5*L1)")
plt.title("Turnover Over Time")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(n_eff, label="Effective # Positions (N_eff)")
plt.title("Portfolio Concentration Over Time")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(11, 5))
plt.stackplot(W.index, [W[c].values for c in W.columns], labels=W.columns)
plt.title("Portfolio Weights (Stacked)")
plt.legend(loc="upper left", ncol=5)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 2.8))
plt.plot(reb_series.index, reb_series.values, linewidth=1)
plt.title("Rebalance Days (1 = Rebalanced)")
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.show()
```
